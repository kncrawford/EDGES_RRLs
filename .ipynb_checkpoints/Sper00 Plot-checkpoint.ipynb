```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
y = open('sper00_col.csv')
reader = csv.reader(y,  delimiter=' ')

data=[] #loads data into array for processing
for row in reader:
    data.append(row)
data=np.array(data)
#print(data)
```

```python
day1 = list(filter(lambda x: x[2]=="1" and x[3]=="2015:287", data))

freq = list(map(lambda x: x[0], day1))
spec = list(map(lambda x: x[1], day1))
```

```python
plt.plot(freq, spec, label='2016:287')

#plt.title("", fontsize='16')
plt.legend()
plt.xlabel("Frequency (MHz)")
plt.ylabel("Spectrum Temperature (K)")
plt.show(block=False)
```
